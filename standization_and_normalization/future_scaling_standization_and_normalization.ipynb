```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load _titanic dataset
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df2=df[["survived","pclass","age","parch","sibsp"]]
df2.head()
```

```python
df3=df2.fillna(df2.mean())
df3.head()
```

```python
x=df3.drop(columns="survived",axis=1)  # survived is a dependent furture
y=df3["survived"]
print("shape of x",x.shape)
print("shape of y",y.shape)
```

```python
x.head()
```
